```python
def gaussian(x, mu, sigma):
    # This einsum call calculates (x-mu)T.Sigma-1.(x-mu) in a vectorized
    # way across all the input variables.
    Sigma_det = np.linalg.det(sigma)
    Sigma_inv = np.linalg.inv(sigma)
    fac = np.einsum('...k,kl,...l->...', x-mu, Sigma_inv, x-mu)
    return 2*np.exp(-fac / 2) / 2
```

```python
X = np.linspace(-10,10,500)
Y = np.linspace(-10,10,500)
X, Y = np.meshgrid(X, Y)
pos = np.empty(X.shape + (2,))
pos[:, :, 0] = X
pos[:, :, 1] = Y
z1 = gaussian(pos, [-5., -5.], np.array([[1. , 0.5], [0.5,  1.]]))
z2 = gaussian(pos, [5., 0.], np.array([[ 5. , 2], [2,  5.]]))
#Make a 3D plot
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface(X, Y, z1,cmap='hot',linewidth=0,antialiased=True)
ax.plot_surface(X, Y, z2,cmap='hot',linewidth=0,antialiased=True)
plt.show()
```
