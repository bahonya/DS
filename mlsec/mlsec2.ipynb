```python
def gaussian(x, y, mu, sigma):
    return np.exp(-((x-mu[0])**2/2*sigma[0]**2) + ((y-mu[1])**2/2*sigma[1]**2))
```

```python
X = np.linspace(-10,10,500)
Y = np.linspace(-10,10,500)
X, Y = np.meshgrid(X, Y)
#z1 = gaussian(X, Y, [-5, -5], [0.5, 0.5])
z2 = gaussian(X, Y, [5, 0], [2, 2])
print(z2)
#Make a 3D plot
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
#ax.plot_surface(X, Y, z1,cmap='hot')
ax.plot_surface(X, Y, z2,cmap='hot')
plt.show()
```

```python
def meshgrid(n):
  X = np.linspace(0, n, n)
  Y = np.linspace(0, n, n)
  X, Y = np.meshgrid(X,Y)
  return X, Y
```

```python
def in_circle(X, Y, r):
  return (np.sqrt((X-X.shape[0]/2)**2 + (Y-Y.shape[0]/2)**2) < r).astype(int)
```

```python
def check_meshgrid_circle(n, r):
    X, Y = meshgrid(n)
    return in_circle(X, Y, r)
```

```python
plt.imshow(check_meshgrid_circle(5, 2.7), cmap='Greys')
```
